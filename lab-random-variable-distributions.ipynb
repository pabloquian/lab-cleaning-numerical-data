```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset
numerical = pd.read_csv('numerical.csv')
numerical.head()
```

```python
numerical['Unnamed: 0'].nunique
```

```python
numerical = numerical.drop(['Unnamed: 0'], axis=1)
```

```python
for col in numerical.columns:
    sns.displot(numerical[col])
    plt.show()
```

```python
log_numerical = np.log(numerical)
# This transformation help us to reshape the data more like a normal distribution.
# Some models prefer data to be more like a normal distribution.
log_numerical.head()
```

```python
for col in log_numerical.columns:
    sns.displot(log_numerical[col])
    plt.show()
```
